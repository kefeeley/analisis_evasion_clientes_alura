```python
# Importación de librerías necesarias
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore') # Evita mensajes de advertencia visuales

# ==============================================================================
# PARTE 1: EXTRACCIÓN DE DATOS
# ==============================================================================
print("="*60)
print(" PARTE 1: EXTRACCIÓN DE DATOS ")
print("="*60)

url_api = "https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/main/TelecomX_Data.json"
respuesta = requests.get(url_api, timeout=10)
respuesta.raise_for_status()
df_telecom = pd.json_normalize(respuesta.json())

print("RESULTADO ESPERADO PARTE 1:")
print(f"-> Conexión exitosa. Se extrajeron {df_telecom.shape[0]} filas y {df_telecom.shape[1]} columnas.")
print("-> Muestra de datos EN CRUDO (nota los nombres de columnas anidados):")
display(df_telecom.head(2))


# ==============================================================================
# PARTE 2: TRANSFORMACIÓN Y LIMPIEZA
# ==============================================================================
print("\n" + "="*60)
print(" PARTE 2: TRANSFORMACIÓN Y LIMPIEZA ")
print("="*60)

# 1. Limpieza Estructural
for prefijo in ['customer.', 'phone.', 'internet.', 'account.', 'Charges.']:
    df_telecom.columns = df_telecom.columns.str.replace(prefijo, '', regex=False)
df_telecom.columns = df_telecom.columns.str.lower()

# 2. Tratamiento de Tipos y Nulos
df_telecom['total'] = pd.to_numeric(df_telecom['total'], errors='coerce').fillna(0.0)

# 3. Traducción y Estandarización
traducciones = {'Yes': 'Sí', 'No': 'No', 'Male': 'Masculino', 'Female': 'Femenino',
                'No internet service': 'Sin internet', 'No phone service': 'Sin teléfono',
                'Month-to-month': 'Mensual', 'One year': '1 Año', 'Two year': '2 Años',
                'Bank transfer (automatic)': 'Transferencia', 'Credit card (automatic)': 'Tarjeta',
                'Electronic check': 'Cheque electrónico', 'Mailed check': 'Cheque por correo'}
df_telecom = df_telecom.replace(traducciones)

# 4. Filtro de Calidad en Variable Objetivo
registros_iniciales = len(df_telecom)
df_telecom = df_telecom[df_telecom['churn'].isin(['Sí', 'No'])].copy()

# 5. Feature Engineering
etiquetas = ['0-1 Año', '1-2 Años', '2-3 Años', '3-4 Años', '4-5 Años', 'Más de 5 Años']
df_telecom['rango_permanencia'] = pd.cut(df_telecom['tenure'], bins=[-1, 12, 24, 36, 48, 60, np.inf], labels=etiquetas)
df_telecom['churn_num'] = df_telecom['churn'].map({'Sí': 1, 'No': 0})

print("RESULTADO ESPERADO PARTE 2:")
print(f"-> Registros anómalos eliminados: {registros_iniciales - len(df_telecom)}")
print("-> Tipos de datos corregidos ('total' ahora es float64 y no hay nulos):")
print(df_telecom[['tenure', 'monthly', 'total']].info())
print("\n-> Muestra de datos LIMPIOS con nuevas columnas (rango_permanencia y churn_num):")
display(df_telecom[['customerid', 'contract', 'total', 'churn', 'churn_num', 'rango_permanencia']].head(3))


# ==============================================================================
# PARTE 3: CARGA Y ANÁLISIS EXPLORATORIO (EDA)
# ==============================================================================
print("\n" + "="*60)
print(" PARTE 3: ANÁLISIS EXPLORATORIO (EDA) ")
print("="*60)
print("RESULTADO ESPERADO PARTE 3:")
print("-> Renderizando panel visual con 4 gráficos de diagnóstico...")

sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Gráfico 1
ax1 = sns.countplot(data=df_telecom, x='churn', hue='churn', palette='Set2', legend=False, ax=axes[0, 0])
axes[0, 0].set_title('1. Tasa General de Evasión (Churn)', fontweight='bold')
for p in ax1.patches:
    ax1.annotate(f'{100 * p.get_height() / len(df_telecom):.1f}%',
                 (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')

# Gráfico 2
sns.countplot(data=df_telecom, x='contract', hue='churn', palette='Set2', ax=axes[0, 1])
axes[0, 1].set_title('2. Impacto de los Tipos de Contrato', fontweight='bold')

# Gráfico 3
sns.countplot(data=df_telecom, x='rango_permanencia', hue='churn', palette='Set2', ax=axes[1, 0])
axes[1, 0].set_title('3. Evasión por Años de Permanencia', fontweight='bold')
axes[1, 0].tick_params(axis='x', rotation=45)

# Gráfico 4
sns.kdeplot(data=df_telecom, x='monthly', hue='churn', fill=True, palette='Set2', common_norm=False, ax=axes[1, 1])
axes[1, 1].set_title('4. Sensibilidad al Precio Mensual', fontweight='bold')

plt.tight_layout()
plt.show()


# ==============================================================================
# PARTE 4: INFORME FINAL
# ==============================================================================
print("\n" + "="*60)
print(" PARTE 4: INFORME EJECUTIVO ")
print("="*60)
print("RESULTADO ESPERADO PARTE 4:")
print("-> Impresión del diagnóstico automatizado para entrega a negocio:")

print("""
[ DIAGNÓSTICO DE RETENCIÓN - TELECOM X ]
- Tasa Crítica: El 26.5% de la cartera de clientes válidos ha abandonado la empresa.
- Fuga por Contrato: La modalidad 'Mensual' concentra la inmensa mayoría de las deserciones.
- Riesgo Temprano: El primer año (0-12 meses) es el período de mayor vulnerabilidad.
- Factor Precio: Se detecta alta evasión en el segmento de tarifas premium ($70-$100).

[ RECOMENDACIONES ]
1. Incentivar agresivamente la migración de planes mensuales a anuales.
2. Desplegar un programa de fidelización intensivo durante el primer año.
3. Revisar la propuesta de valor y calidad de los servicios más costosos.
""")
```
